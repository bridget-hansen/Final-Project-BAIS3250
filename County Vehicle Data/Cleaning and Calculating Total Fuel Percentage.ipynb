# Bridget Hansen - Cleaning Vehicle Data and Calculating Total Fuel Percentage

```python
import pandas as pd
```

## Reading in the Excel File from Data.gov

```python
#Reading in the Excel File
county_vehicles_df = pd.read_excel("2016cityandcountylightdutyvehicleinventory.xlsx", header=1, sheet_name=3)
county_vehicles_df
```

```python
#Checking column names 
county_vehicles_df.columns
```

## Data Reduction and Cleaning 

```python
#Only keeping necessary columns
columns_to_keep = ['state_abbr', 'county_id', 'county_name', 'fuel_type_org', 'before 1980', '1980-99', '1990-99',2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
county_vehicles_reduced = county_vehicles_df[columns_to_keep]
county_vehicles_reduced
```

```python
#Changing the data types
county_vehicles_reduced["state_abbr"] = county_vehicles_reduced["state_abbr"].astype("category")
county_vehicles_reduced["county_name"] = county_vehicles_reduced["county_name"].astype("category")
county_vehicles_reduced["fuel_type_org"] = county_vehicles_reduced["fuel_type_org"].astype("category")

# verifying the data types
print(county_vehicles_reduced.info())
```

```python
#Finding the na values
county_vehicles_reduced.isna().sum()
```

```python
#Coverting all column names to strings instead of integers
county_vehicles_reduced.columns = county_vehicles_reduced.columns.astype(str)
```

```python
#Filling Na values only in the year columns
numeric_cols = county_vehicles_reduced.select_dtypes(include=['float64']).columns
county_vehicles_reduced[numeric_cols] = county_vehicles_reduced[numeric_cols].fillna(0)
county_vehicles_reduced
```

## Calculating Total Electric Vehicle Percentage

```python
#Cleaning the fuel type orginal column 
county_vehicles_reduced['fuel_type_org'] = county_vehicles_reduced['fuel_type_org'].str.strip().str.upper()

#Defining the vehicle types to keep
ev_types = ['ELECTRIC VEHICLE', 'HYBRID ELECTRIC VEHICLE', 'PLUG IN HYBRID ELECTRIC VEHICLE']

#filtering to only include electric vehicles
ev_only = county_vehicles_reduced[county_vehicles_reduced['fuel_type_org'].isin(ev_types)].copy()
```

```python
#Displaying the data frame to verify
ev_only.head(5)
```

```python
#Dropping fuel type column since data frame was reduced to only electric
ev_only = ev_only.drop(columns=['fuel_type_org'])
ev_only
```

```python
# Identifying all the year columns
year_cols = [col for col in ev_only.columns if col.startswith('before') or col.isdigit()]

# Summing up the ev percentages in all rows
ev_only['ev_total_row'] = ev_only[year_cols].sum(axis=1)
```

```python
#Displaying the dataframe to verify
ev_only.head(5)
```

## Creating a New DataFrame with County Id and Total EV Percentage

```python
#Summing up the total ev_row for each county to get a total percentage of all ev types
total_ev_percentage = ev_only.groupby('county_id')['ev_total_row'].sum().reset_index(name='total_ev_percentage')
total_ev_percentage
```

```python
# Keep the first instance of each row based on state_abbr, county_id, and county_name to not have repeat entries 
ev_only_unique = ev_only.drop_duplicates(subset=['state_abbr', 'county_id', 'county_name'])
ev_only_unique
```

```python
#Defininf only the columns necesseary to keep and dropping all the year columns
ev_only_unique = ev_only_unique[['state_abbr', 'county_id', 'county_name']]
ev_only_unique
```

## Merging Cleand EV only DataFrame with The total EV Percentage 

```python
#merging the data frames to include the total ev percentage
county_fuel = ev_only_unique.merge(total_ev_percentage, on='county_id', how='left')
county_fuel
```

```python
# Multiply each value in the total ev_percentage column by 100 to convert it from decimal to percent
county_fuel['total_ev_percentage'] = county_fuel['total_ev_percentage'] * 100
county_fuel
```

```python
#Renaming columns
county_fuel.rename(columns={'state_abbr': 'state', 'county_name' : 'county', 'total_ev_percentage':'electric vehicle %'}, inplace=True)
county_fuel
```

## Saving to a CSV File

```python
#Saving the cleaned data set to a CSV file
county_fuel.to_csv("county_fuel.csv", index=False, encoding="utf-8")
```
